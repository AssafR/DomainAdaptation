```python
import os
import shutil
from pathlib import Path
```

```python
classes = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
sexes = {'M':'Male','F':'Female','U':'Undecidable'}
srcDir = Path("C:\Labs") / "DataWithGender"
dstDir = Path("C:\Labs") / "DataDomains"
```

```python
list = os.listdir(srcDir/"Angry_F") # dir is your directory path
number_files = len(list)
print (number_files)
```

```python
"""
Dir architecture: 
    - train: Men, Female, Undecidable
    - val: Men, Female, Undecidable

"""
for sex in sexes.keys():
    for dir_type in ['train', 'val']:
        new_dir = dstDir / dir_type / sexes[sex]
        if not new_dir.exists():
            new_dir.mkdir(parents=True,exist_ok=True)

total=0            
for sex in sexes.keys():
    for emotion in classes:
        dirName = f"{emotion}_{sex}"
        curr_dir = srcDir / dirName
        number_files = len(os.listdir(curr_dir))

        transferred_files = 0
        for srcFile in curr_dir.iterdir():
            if transferred_files <= number_files*0.8:
                targetFile = dstDir / "train" / sexes[sex]/srcFile.name
                if not targetFile.exists():
                    print("Copying file ",srcFile.resolve()," to ", targetFile.resolve())
                    shutil.copyfile(srcFile.resolve(), targetFile.resolve(), follow_symlinks=False)
                    transferred_files += 1
            else: 
                targetFile = dstDir / "val" / sexes[sex]/srcFile.name
                if not targetFile.exists():
                    print("Copying file ",srcFile.resolve()," to ", targetFile.resolve())
                    shutil.copyfile(srcFile.resolve(), targetFile.resolve(), follow_symlinks=False)
                    transferred_files += 1

        total += transferred_files
                
print(f"Total {total} files copied")
```
